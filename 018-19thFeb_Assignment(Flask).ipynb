## Q1. What is Flask Framework? What are the advantages of Flask Framework? 

- Flask is a micro web framework written in Python that provides a simple and easy way to build web applications.
####
- Advantages of Flask
    - Lightweight
        - Flask is designed to be a lightweight and simple to use framework for building web applications.
    - Ease to use
        -It provides a minimal set of tools to get started quickly, making it a great choice for small projects or for developers who want to get up and running quickly.
    - Easy to learn
        -  Flask is a micro framework that has a simpler and more straightforward design compared to full-stack frameworks that makes it easier to learn and use for beginners.

## Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# Flask application to display 'Hello World!!'

# <-- CODE -->

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0", port=8080)

# URl for Output- https://mango-artist-naczq.pwskills.app:8080/
```

## Q3. What is App routing in Flask? Why do we use app routes?

- The process of defining the URLs that map to specific functions or methods in your application.
- We use app routes to determine which function or method should handle the request and return the response when a user visits a URL in our application

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

- url_for() function is used in Flask for URL binding.
- The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL.

```python
from flask import *

app = Flask(__name__)

@app.route('/hello')
def hello():
    return 'Hello World!'

@app.route('/hello1')
def hello1():
    return 'Hello World_1!'


@app.route('/test/<name>')
def user(name):
    if name == 'hello':
        return redirect(url_for('hello'))
    if name == 'hello1':
        return redirect(url_for('hello1'))


if __name__ == '__main__':
     app.run(host="0.0.0.0", port=8080)
```
