## Q1. Explain GET and POST methods.

- GET
    - Data is sent through urlbar.
    - Only limited amount of data can be sent as data is sent in header.
    - Therequest is not secured because data is exposed in URL bar.
    - The request can be bookmarked.
    - Get request is idempotent. It means second request will be ignored until response of first request is delivered.
    - Get request is more efficient and used more than Post.
####
- POST
    - Data is sent as body.
    - Large amount of data can be sent as data is sent in body.
    - The request is secured because data is not exposed in URL bar.
    - The request cannot be bookmarked.
    - The request is non-idempotent. It means multiple POST requested can be addressed simultaneously.
    - Post request is less efficient and used less than get.

## Q2. Why is request used in Flask?

- 'request' is used in Flask to handle a request which creates a Request object based on the environment it received from the WSGI server.
- 'request' contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods.

## Q3. Why is redirect() used in Flask? 

- redirect() function returns a response object and redirects the user to another target location with specified status code.
    - e.g return redirect("/home", code=302)

## Q4. What are templates in Flask? Why is the render_template() function used? 

- Templates are files that contain static data as well as placeholders for dynamic data.
####
- render_template() function is used to produce a final document from a template that is being rendered with some specific data.
    e.g return render_template('index.html')
- Flask uses the Jinja template library to render HTML templates.
