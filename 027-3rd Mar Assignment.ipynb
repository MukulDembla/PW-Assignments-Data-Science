## Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot. 

1. Line Plot: A line plot can be created using the sns.lineplot() function, which plots a line connecting the data points in a dataset.
    - This type of plot is useful for visualizing trends and changes over time.
####
2. Scatter Plot: A scatter plot can be created using the sns.scatterplot() function, which plots individual data points as dots on a 2D plane.
    - This type of plot is useful for visualizing the relationship between two variables.
####
3. Histogram: A histogram can be created using the sns.histplot() function, which shows the distribution of a single variable by grouping the data into bins and counting the number of data points in each bin.
    - This type of plot is useful for understanding the shape and spread of a dataset.
####
4. Box Plot: A box plot can be created using the sns.boxplot() function, which shows the distribution of a dataset by displaying the quartiles, median, and outliers.
    - This type of plot is useful for comparing the distributions of multiple datasets.
####
5. Pair Plot: A Pair plot can be created using the sns.pairplot() function creates a grid of scatter plots and histograms, where each variable is plotted against every other variable in the dataset.
    - The Pair Plot is particularly useful for identifying patterns and relationships between variables, as well as for detecting outliers and anomalies.

## Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
#### Note:  timepoint, signal, event, and region are columns in the fmri dataset. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, hue = fmri.event, style = fmri.region)
```

## Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
#### Note:  pclass, age, and fare are columns in the titanic dataset. 

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.age)
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.fare)
```

## Q4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
sns.histplot(x='price', hue = 'cut', data = diamonds, multiple = 'stack')
```

##  Q5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris, hue = 'species')
```

## Q6. Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot_table(index = 'year', columns = 'month', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt='d')
```
