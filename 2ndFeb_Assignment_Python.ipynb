### Solution 1

```python
# For Loop - We use for loop when no of iterations are not known.
#          - When condition for breaking the loop is not specified.

# Printing 1 to 9
for i in range(10): 
    print(i)
```

```python
# While Loop - We use while loop when the no of iterations are known.
#            - When loop has to be breaked if a condition is fulfilled.

# Printing 0 to 9
i = 0
while (i < 10):  # Here i greater than 0 is a condition
    print(i)
    i += 1
```

### Solution 2

```python
sum = 0
product = 1
for i in range(1,11):
    sum += i
    product *= i
print(f'The sum of first 10 natural number is: {sum}')
print(f'The product of first 10 natural number is: {product}')
```

### Solution 3

```python
units = int(input('Enter Units: '))
bill = 0
if units <= 100:
    bill += 4.5*units
    units -= 100
elif units > 100 and units <= 200:
    bill += 450
    units -= 100
    bill += 6*units
elif units > 200 and units <= 300:
    bill += 450+600
    units -= 200
    bill += 10*units
elif units > 300:
    bill += 450+600+1000
    units -= 300
    bill += 20*units
print(f'Total Electricity Bill Rs.{bill}')
```

### Solution 4

```python
lst = list(range(1,101))

#using for loop
l1 = []
for i in lst:
    c = i**3
    if c%4==0 or c%5==0:
        l1.append(i)
print(l1)
```

```python
#using while loop
l2 = []
i = 1
while i <= len(lst):
    c = i**3
    if c%4==0 or c%5==0:
        l2.append(i)
    i += 1
print(l2)
```

### Solution 5

```python
string = 'I want to become a data scientist'
count = 0
vowel = []
for i in string:
    if i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i=='A' or i=='E' or i=='I' or i=='O' or i=='U':
        count += 1
        vowel.append(i)
print(f'Total number of vowels in the given string are: {count}')
```
