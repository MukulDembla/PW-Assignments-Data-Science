```python
class vehicle:
    
    def __init__(self, name, max_speed, average):
        self.name = name
        self.max_speed = max_speed
        self.average = average
```

```python
class car(vehicle):
    
    def seating_caacity(self, capacity):
        self.capacity = capacity
        return f'The {self.name} has a seating capacity of {self.capacity}'
```

```python
c = car('tata',120, 15)
c.seating_caacity(5)
```

- Multiple inheritance - When child class inherits from more than one parent class.

```python
class data_science:
    def student_details_ds(self):
        return 'this meth give student details from data science batch'
    
class web_dev:
    def student_details_wd(self):
        return 'this meth give student dedtails from web dev batch'
    
class student(data_science,web_dev):
    def __init__(self,name):
        self.name = name
```

```python
st = student('mukul')
```

```python
st.student_details_wd()
```

```python
st.student_details_ds()
```

- In Python, a getter is a method used to retrieve the value of an instance variable, while a setter is a method used to set the value of an instance variable. Getters and setters provide an additional layer of abstraction that helps to protect the data of the class from direct access and modification.

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self._age = age
    
    def get_age(self):
        return self._age
    
    def set_age(self, age):
        self._age = age
```

```python
prsn = person('mukul',21)
```

```python
prsn.get_age()
```

```python
prsn.set_age(20)
```

```python
prsn.get_age()
```

- Method overriding in Python is to create two methods with the same name and same parameter in a superclass-subclass relationship.

```python
class pwskills:
    def course(self):
        print('courses offered by pw skills')
        
class ds(pwskills):
    def course(self):
        print('this student is from data science batch')
        
class wd(pwskills):
    def course(self):
        print('this student is from web devloment batch')
```

```python
pw = pwskills()
mukul = ds()
sudh = wd()
```

```python
pw.course()
mukul.course()
sudh.course()
```
