```python
from functools import reduce
```

```python
def product_of_numbers(lst):
    list_1 = []
    for item in lst:
        if type(item) == int or type(item) == float:
            list_1.append(item)
        elif type(item) == tuple:
            for i in item:
                if type(i) == int or type(i) == float:
                    list_1.append(i)
        elif type(item) == list:
            for i in item:
                if type(i) == int or type(i) == float:
                    list_1.append(i)
        elif type(item) == set:
            for i in item:
                if type(i) == int or type(i) == float:
                    list_1.append(i)
        elif type(item) == dict:
            for key,value in item.items():
                if type(key) == int or type(key) == float:
                    list_1.append(key)
                if type(value) == int or type(value) == float:
                    list_1.append(value)
                elif type(value) == tuple or type(value) == list:
                    for i in value:
                        if type(i) == int or type(i) == float:
                            list_1.append(i)
                            

    product = reduce(lambda x,y: x*y, list_1) 
        
    return product
```

```python
ques_list = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

result = product_of_numbers(ques_list)
print(result)
```

```python
def encrypt_sentence(sentence):
    sentence = sentence.lower()
    encrypted_sentence = ""
    for char in sentence:
        if char.isalpha():
            encrypted_char = chr(ord('a') + (ord(char) - ord('a') + 25) % 26)
        elif char == " ":
            encrypted_char = "$"
        else:
            encrypted_char = char
        encrypted_sentence += encrypted_char
    return encrypted_sentence

sentence = "I want to become a Data Scientist."
encrypted_sentence = encrypt_sentence(sentence)
print("Encrypted Sentence:", encrypted_sentence)
```
